## –ó–∞–≤–¥–∞–Ω–Ω—è 5 (```if ... elif ... else```)

**–£–º–æ–≤–∞:**

–í–≤–µ—Å—Ç–∏ –æ—Ü—ñ–Ω–∫—É —Å—Ç—É–¥–µ–Ω—Ç–∞ (—Ü—ñ–ª–µ —á–∏—Å–ª–æ –≤—ñ–¥ ```0``` –¥–æ ```100```).

–í–∏–≤–µ—Å—Ç–∏ —Ä—ñ–≤–µ–Ω—å –∑–Ω–∞–Ω—å:

* ```90‚Äì100``` ‚Üí –í—ñ–¥–º—ñ–Ω–Ω–æ
* ```75‚Äì89``` ‚Üí –î–æ–±—Ä–µ
* ```60‚Äì74``` ‚Üí –ó–∞–¥–æ–≤—ñ–ª—å–Ω–æ
* –º–µ–Ω—à–µ ```60``` ‚Üí –ù–µ–∑–∞–¥–æ–≤—ñ–ª—å–Ω–æ

````python
#### Variant ‚Ññ1 ####

# Input data
grade = 95

# Solution
if grade < 60:
    result = "–ù–µ–∑–∞–¥–æ–≤—ñ–ª—å–Ω–æ"
elif grade <= 74:
    result = "–ó–∞–¥–æ–≤—ñ–ª—å–Ω–æ"
elif grade <= 89:
    result = "–î–æ–±—Ä–µ"
elif grade <= 100:
    result = "–í—ñ–¥–º—ñ–Ω–Ω–æ"
else:
    result = ""

# Print results
print(result)
````
